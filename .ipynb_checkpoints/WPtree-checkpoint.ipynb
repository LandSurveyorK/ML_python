```python
import numpy as np
import matplotlib.pyplot as plt
import scipy 
import pandas as pd
```

```python
def feature( n ):
    X = np.random.uniform(low=0,high=1,size=(n,2))
    return X

X = feature(1000)

def response1(X,sigma):
    U1 = 2*(X[:,0]< 0.5)-1
    V1 = 2*(X[:,1] < 0.5)-1
    noise = np.random.normal(0,sigma,len(X))
    Y = U1 * V1 + noise
    return Y

n = 1000
X = feature(1000); Y = response1(X,0.2)
dataMat = np.empty([n,3]); dataMat[:,:-1] =  X; dataMat[:,-1] = Y 

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.scatter(X[:,0],X[:,1], Y, c = 'b', marker = '.')
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('Y')

plt.show()
```

```python
            
def bestSplitting(dataMat,M_try):
    X = dataMat[:,:-1]; y = dataMat[:,-1]
    feat_index = 0; location = 0; reducement = 0
    for i in M_try:
        sort_on_i = np.sort(X[:,i])
        for value in sort_on_i:
            
            right_dataMat = dataMat[dataMat[:,i] > value]
            n2 = len(right_dataMat)
            left_dataMat = dataMat[dataMat[:,i] <= value]
            n1 = len(left_dataMat)

            new_reducement = np.var(dataMat[:,-1]) - 1.0 * n1 / n * np.var(left_dataMat[:,-1]) - 1.0 * n2 / n * np.var(right_dataMat[:,-1])
            if reducement < new_reducement:
                reducement = new_reducement
                feat_index = i; location = value
    
    l_data = dataMat[dataMat[:,feat_index] <= location]; n1 = len(l_data);l_mse = 1.0 * n1 / n * np.var(l_data)
    r_data = dataMat[dataMat[:,feat_index] > location]; n2 = len(r_data); r_mse = 1.0 * n2 / n * np.var(r_data)
    
    return l_data, r_data, l_mse, r_mse,n1,n2,feat_index, location
```

```python

max_leaf = 10

def consistTree(dataMat):

    n, num_feat = dataMat[:,:-1].shape
    mse = np.var(dataMat[:,-1]); tree = {'root':{'Node_member': dataMat,'mse' : mse, 'num_observation': n }}
    M = int(np.sqrt(num_feat) )
    M_try = np.random.choice(num_feat, M)
    
    treeTrack =[]; num_split = 0
    new_tree = tree
    while(num_split < max_leaf ):
    
        for  key  in list(tree):
            data_to_split = tree[key]['Node_member']
            l_data, r_data, l_mse, r_mse,n1,n2,feat_index, location = bestSplitting(data_to_split, M_try)
            tree[key]['feat_index'] = feat_index
            tree[key]['location'] = location
            treeTrack.append(tree) 
            
            del new_tree[key]
            new_tree[str(num_split) + 'l_leaf'] = {'Node_member': l_data, 'mse': l_mse, 'num_observation': n1}
            new_tree[str(num_split) + 'r_dleaf'] = {'Node_member': r_data, 'mse': r_mse, 'num_observation': n2}
            num_split += 1 
            if num_split > max_leaf: 
                tree = new_tree; break
        tree = new_tree
                  
    last_tree = treeTrack[-1]; terminal_leaves = {} 
    
    for key in list(last_tree):
        terminal_leaves[key] ={
            'mse': last_tree[key]['mse'],
         #   'feat_index': last_tree[key]['feat_index'],
          #  'location' : last_tree[key]['location'],
            'num_observation': last_tree[key]['num_observation']}     
    
    return treeTrack, terminal_leaves


treeTrack, terminal_leaves = consistTree(dataMat)
for key in list(terminal_leaves):
    print(terminal_leaves[key], end ="")
    print('\n')
    
```
