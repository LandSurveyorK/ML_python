```python
import numpy as np

n = 100;
x = []; y = []; z = []
for i in range(100):
    for j in range(100):
        x.append(-1 + 0.02 * i )
        y.append(-1 + 0.02 * j )
        z.append((-1 + 0.02 * i ) * (-1 + 0.02 * j ))

print("done!")
import matplotlib.pyplot as plt
plt.ion()

from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.scatter(x,y,z, c = 'b', marker = 'o')
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('Y')
plt.show()
```
